# Análisis del CSV de ejemplo

Este notebook reproduce el análisis del Módulo 1 en formato Jupyter. En el Módulo 2 vamos a migrarlo a un script `.py` modular y reutilizable.

**Requisitos:**
- Tener `datos_ejemplo.csv` en la misma carpeta (generado con `generar_datos.py`)
- Tener pandas y openpyxl instalados en el entorno virtual

## 1. Importar librerías

```python
import pandas as pd
```

## 2. Cargar los datos

```python
df = pd.read_csv("../datos_ejemplo.csv")
df.head()
```

## 3. Exploración rápida

Resumen estadístico de las variables numéricas.

```python
df.describe()
```

## 4. Conteo por sexo

```python
df["sexo"].value_counts()
```

## 5. Ingreso promedio por sexo

```python
df.groupby("sexo")["ingreso"].mean().round(0)
```

## 6. Ingreso promedio por sexo y área

```python
df.groupby(["sexo", "area"])["ingreso"].mean().round(0)
```

## 7. Edad promedio por área

```python
df.groupby("area")["edad"].mean().round(1)
```

## 8. Exportar el resumen a Excel

```python
resumen = df.groupby(["sexo", "area"])["ingreso"].mean().round(0).reset_index()
resumen.columns = ["sexo", "area", "ingreso_promedio"]
resumen.to_excel("resumen.xlsx", index=False)
print("Archivo resumen.xlsx creado.")
```

## Observaciones

Este notebook funciona como exploración. Tiene varios problemas si pensamos en producción:

- No es directamente reutilizable: si cambias el archivo, hay que editar varias celdas
- No tiene funciones — el código se repite (`.groupby()...mean().round()` aparece varias veces)
- Para correrlo en otro contexto (programado, automatizado) hay que reescribir todo
- Difícil de revisar en Git: el `.ipynb` es JSON con muchos metadatos que ensucian los diffs

En el Módulo 2 vamos a migrarlo a un `.py` que resuelve todos estos problemas.
